```python
#Pulling 201701 Data
Data_201701 = pd.read_csv('201701-citibike-tripdata.csv')
Data_201701.head()
```

```python
Data_201701.columns
```

```python
# Rename Columns
Data_201701 = Data_201701.rename(columns={'Trip Duration': 'tripduration', 'Start Time': 'starttime','Stop Time':'stoptime', 
                                          'Start Station ID' : 'start station id','Start Station Name' : 'start station name' ,
                                          'Start Station Latitude': 'start station latitude' ,'Start Station Longitude':'start station longitude',
                                          'End Station ID' : 'end station id' , 'End Station Name':'end station name',
       'End Station Latitude': 'end station latitude', 'End Station Longitude':'end station longitude', 'Bike ID':'bikeid',
                                          'User Type':'usertype' ,'Birth Year': 'birth year', 'Gender':  'gender'  })


```

```python
#Removing Extra columns
Data_201805 = Data_201805.drop(["name_localizedValue0"],axis=1)
Data_201805.head()
```

```python
Data_201801 = Data_201801.drop(["name_localizedValue0"],axis=1)
Data_201801.head()
```

```python
Data_201801.columns
```

```python
#Merging files
#Stack data 
CitiBikeData = pd.concat([Data_201805,Data_201801,Data_201707,Data_201701],axis=0)
#CitiBikeData = pd.merge(Data_201805,Data_201801)

```

```python
CitiBikeData.head(20)
```

```python
#save to csv
CitiBikeData.to_csv('CitiBikeData.csv', index=False)
```
